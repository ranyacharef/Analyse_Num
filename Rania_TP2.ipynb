## Version Python

````python
from numpy import prod
def prog(x,y,c):
    S = [] 
    n = len(x)
    
    for j in range(n):
        L =[c-xx for xx in x]
        for k in range(n):
            if (k == j):                
                L[k] = 1
            else:
                L[k] = L[k]/(x[j] - x[k])   
        S.append(y[j]*prod(L))
        
    return  sum(S)
````

````python

X=[1 ,2 ,3 ,4,-2];
Y=[-1, 0, 2 ,1,3];
for i in X:
    print(prog(X,Y,i))
sum(X)
````

**Polynome d'interpolation de Lagrange** 
> Soient $(x_0,y_0), \ldots,(x_k,y_k),\ldots ,(x_n,k_n)$ $n+1$ points deux à deux distincts, le polynôme d'interpolation de Lagrange associés à ces points supports est défini par :
> $$P_n(x)=\sum_{k=0}^{n-1} y_kL_k(x)$$

avec $L_{0}(x)=\displaystyle\frac{(x-x_1)(x-x_2)\ldots(x-x_{n})}{(x_0-x_1)(x_0-x_2)\ldots(x_0-x_{n})}$ et $L_{k}(x)=\displaystyle\frac{(x-x_1)(x-x_2)\ldots(x-x_{k-1})(x-x_{k+1})\ldots(x_0-x_{n})}{(x_k-x_0)(x_k-x_1)\ldots(x_k-x_{k-1})(x_k-x_{k+1})\ldots(x_k-x_{n})}$ pour $k\in \{1,\ldots,n\}$

1) Que renvoie la fonction PR lorsqu'on lui déclare  **x=(x(1),x(2),...,x(n))x=(x(1),x(2),...,x(n))**  et  **y=(y(1),y(2),...,y(n))y=(y(1),y(2),...,y(n))**  deux vecteurs de même longueur  `n` et `c`  un nombre réel donné.

````python
from numpy import prod
def PR(x,y,c):
    S = [] 
    n = len(x)
    
    for j in range(n):
        L =[c-xx for xx in x]
        for k in range(n):
            if (k == j):                
                L[k] = 1
            else:
                L[k] = L[k]/(x[j] - x[k])   
        S.append(y[j]*prod(L))
        
    return  sum(S)
````

````python
X=[1 ,2 ,3 ,4,-2]
Y=[-1, 0, 2 ,1,3]
for i in X:
    print(PR(X,Y,i))
sum(X)
````

2) Les points de Tchebycheff sur un intervalle $[a;b]$ sont définie par :
  $$x_k=\dfrac{a+b}{2}+\dfrac{a-b}{2}\cos\left(\frac{k-1}{n-1}\pi\right)\hspace{0.5cm} (k=1,..,n: x_1=a:x_n=b)$$

  Vérifier que sur $[-5; 5]$ ; l'interpolation polynômiale de $f(x) = \dfrac{1}{1 + x^2}$ ; en $N$ points de Tchebychev
est meilleure qu'une  interpolation utilisant $N$ points uniforment espacés (on prendra $N=10, 11, 16, 17$).

````python
from math import *
def tcheb(a,b,n):
    V=[]
    for k in range(n):
        V.append((a+b)/2+((a-b)/2)*cos(((k-1)/(n-1))*pi))
      
    return V
````

````python
tcheb(-5,5,11)
````

````python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(3,figsize=(8,8))
plt.gcf().subplots_adjust(left = 0.2, bottom = 0.2, right = 1.5,
                          top = 0.9, wspace = 0.2, hspace = 0.3)
plt.suptitle("Interpolation Tchebycheff ",size="x-large" )
Yt=[]
N=10
f=lambda x: 1/(1+x**2)
Xt=tcheb(-5,5,N)
for i in Xt:
    Yt.append(f(i))
pt=np.polyfit(Xt,Yt,N-1)
plt.subplot(2,2,1)
plt.plot(t,f(t),'r',Xt,Yt,'ob', t, np.polyval(pt,t),'g')
plt.title("avec N=10")
##########################################
Yt=[]
N=11
f=lambda x: 1/(1+x**2)
Xt=tcheb(-5,5,N)
for i in Xt:
    Yt.append(f(i))
pt=np.polyfit(Xt,Yt,N-1)
plt.subplot(2,2,2)
plt.plot(t,f(t),'r',Xt,Yt,'ob', t, np.polyval(pt,t),'g')
plt.title(" avec N=11")
##########################################
Yt=[]
N=16
f=lambda x: 1/(1+x**2)
Xt=tcheb(-5,5,N)
for i in Xt:
    Yt.append(f(i))
pt=np.polyfit(Xt,Yt,N-1)
plt.subplot(2,2,3)
plt.plot(t,f(t),'r',Xt,Yt,'ob', t, np.polyval(pt,t),'g')
plt.title(" avec N=16")
##########################################
Yt=[]
N=17
f=lambda x: 1/(1+x**2)
Xt=tcheb(-5,5,N)
for i in Xt:
    Yt.append(f(i))
pt=np.polyfit(Xt,Yt,N-1)
plt.subplot(2,2,4)
plt.plot(t,f(t),'r',Xt,Yt,'ob', t, np.polyval(pt,t),'g')
plt.title("avec N=17")
````

1) Déterminer en utilisant les deux codes puis par un calcul à la main le polynôme de Newton pour les données suivantes :


<center>
<table>    
      $\begin{array}{|c|c|c|}
   \hline
   i & x_i  & y_i \\
   \hline
   0 &   -1  &   6 \\
   1 &   0 &  1 \\
   2 &  2  &  3 \\
   3 &  5  &  66 \\
   \hline
\end{array}$  
</table>

````python
import numpy as np
def intN(x,y):
    m=len(x)
    y=np.array(y)
    x=np.array(x)
    for k in range(2,m):
        y[k:m]=(y[k:m]-y[k-1:m-1])/(x[k:m]-x[1:m+1-k])
    return y
````

````python
x=[-1,0,2,5]
y=[6,1,3,66]
f=intN(x,y)
print(f)
````

````python
def intNval(x,y,t):
    m=len(x)
   
    x=np.array(x)
    t=np.array(t)
    y=np.array(y)
    p=y[m-1]*np.ones(np.size(t))
    for k in range(m-1,1,-1):
        p=p*(t-x[k])+y[k]
    return p
````

````python
x=[-1,0,2,5]
y=[6,1,3,66]
t=[0,1,2,3]
f=intNval(x,y,t)
print(f)
````
