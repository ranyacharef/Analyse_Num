```python
def dicho(a,b,g,eps):
    '''version it√©rative '''
    m=(a+b)/2
    err=abs(b-a)
    while err>eps :
        if g(m)==0 :
            break
        elif g(a)*g(m)<0 :
            b=m
        else: a=m
        m=(a+b)/2
        err=abs(b-a)
    return m
    
```

```python
print(dicho(-2,-1.5,g,0.001))
```

```python
def dichow(a,b,f,eps):
    """version tant que qui renvoila solution avec le nombre d'''it√©ration """
    m=(a+b)/2
    err=abs(b-a)
    n=0
    while err>eps :
        n+=1
        if f(m)==0 :
            break
        elif f(a)*f(m)<0 :
            b=m
        else: a=m
        m=(a+b)/2
        err=abs(b-a)
    return m,n
```

>
>2) Montrer que l'√©quation $f(x)=0$ admet une solution unique que l'on peut obtenir en utilisant la m√©thode de Newton √† partir >de $x0=1$.

1)

```python
import numpy as np
import matplotlib.pyplot as plt
def newton(a,eps):
    x1=a
    x2=x1-f(x1)/fprime(x1)
    while abs(x1-x2)>eps :
        x1=x2
        x2=x1-f(x1)/fprime(x1)
    sol=x2
    return (sol)
```

```python
f=lambda x: np.exp(-x)-x
fprime= lambda x: -np.exp(-x)-1
t=np.linspace(0,1,100);
plt.plot(t,f(t))
plt.show()
print ("sol= ",newton(0,0.001))

```
